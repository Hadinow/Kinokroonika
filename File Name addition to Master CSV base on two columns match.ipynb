```python
df2 = df_video_list[['File.name', 'archive_fond_nr', 'archive_archival_document_nr']]
```

```python
df2
```

```python
# create a dictionary mapping (archive_nr, document_nr) tuples to File.name for only matching rows
occ_dict = {}
for _, row1 in df1.iterrows():
    for _, row2 in df2.iterrows():
        if row1['archive_fond_nr'] == row2['archive_fond_nr'] and row1['archive_archival_document_nr'] == row2['archive_archival_document_nr']:
            occ_dict[(row1['archive_fond_nr'], row1['archive_archival_document_nr'])] = row2['File.name']
            break

# add a new 'Occupation' column to df1 and populate it based on the occ_dict
df1['File.name'] = [occ_dict.get((archive_nr, document_nr), None) for archive_nr, document_nr in zip(df1['archive_fond_nr'], df1['archive_archival_document_nr'])]
```
