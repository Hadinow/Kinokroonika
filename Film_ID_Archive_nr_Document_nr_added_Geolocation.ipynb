```python
ID_Location = df1[["film_id", "aggregated_loc", "production_year"]]
```

```python
filtered_locations = []
prefix = 'LOC'
for entities in ID_Location['aggregated_loc']:
    locations_per_film = [] 
    if  entities is not np.nan:
        convert_entities = ast.literal_eval(entities)
        for word in convert_entities:
            if word[0] == prefix:
                locations_per_film.append(word[1])
    filtered_locations.append(locations_per_film)
```

```python
location_filmid = {}

for i, locations in enumerate(filtered_locations):
    film_idx = ID_Location.iloc[i]['film_id']
    for loc in locations:
        if loc in location_filmid:
            location_filmid[loc].append(film_idx)
        else:
            location_filmid[loc] = [film_idx]
            
location_filmid
```

```python
df2['film_ids'] = [location_filmid[x] for x in df2['Place']]
df2
```
