```python
import pandas as pd
```

```python
df1 = pd.read_excel('/Users/hadinowandish/Desktop/2023_01_27_for_processing_video_list.xlsx')
```

```python
df1
```

```python
# find duplicate rows based on two columns at the same time
duplicateRowsDF = df1[df1.duplicated([ 'archive_fond_nr', 'archive_archival_document_nr'], keep=False)]

# create a new dataframe from duplicate rows
if len(duplicateRowsDF) > 0:
    df_duplicates = pd.DataFrame(duplicateRowsDF)
    df_duplicates.sort_values(by=['archive_fond_nr', 'archive_archival_document_nr'], inplace=True)
    print("Dataframe containing only the duplicate rows based on two columns 'archive_archival_document_nr' & 'archive_fond_nr':\n", df_duplicates)
else:
    print("No duplicate rows found based on two columns 'archive_archival_document_nr' & 'archive_fond_nr'.")
```
